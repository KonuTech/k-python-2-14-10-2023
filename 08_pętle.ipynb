W Pythonie, ptla while stanowi pot偶ne narzdzie, pozwalajce na wielokrotne wykonanie bloku kodu do momentu, a偶 okrelony warunek przestanie by speniony. Ptla while w Pythonie dziaa na prostej zasadzie: wykonuje blok kodu tak dugo, jak dugo okrelony warunek jest prawdziwy. Oto podstawowy przykad:

````python
licznik = 1

while licznik <= 5:
    print(f"Numer: {licznik}")
    licznik += 1
````

- 'continue': Powoduje natychmiastowe przejcie do kolejnej iteracji ptli, pomijajc reszt kodu w bie偶cym cyklu.

````python
licznik = 0
while licznik < 10:
    licznik += 1
    if licznik % 2 == 0:
        continue
````

- 'break': Natychmiast przerywa dziaanie ptli, niezale偶nie od warunku logicznego okrelonego w ptli while.

Tutaj, gdy 'licznik' osignie warto 5, instrukcja 'break' przerwie dziaanie ptli, nawet jeli warunek 'licznik < 10' nadal jest speniony.

````python
licznik = 0
try:
    while licznik < 5:
        print(f"Licznik: {licznik}")
        licznik += 1
        if licznik == 3:
            break
finally:
    print("Ptla zostaa zakoczona.")
````
